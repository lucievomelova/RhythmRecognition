# Spectral-based novelty function
Spectral-based novelty function, also known as spectral flux, is computed from time-frequency representation of the audio signal. The idea is that by tracking changes in frequency content of the signal, we can detect note onsets.

```python
import numpy as np
from matplotlib import pyplot as plt
import librosa.display
from RhythmRecognition.constants import *
from RhythmRecognition.onset import spectral

audiofile = "../audio_files/spark.wav"

spectral_novelty = spectral.SpectralNovelty(audiofile)
spectral_novelty = spectral_novelty.get()

frames = np.arange(len(spectral_novelty))
t = librosa.frames_to_time(frames, sr=SAMPLING_RATE, hop_length=HOP_LENGTH)

plt.figure(figsize=(12, 3))
plt.plot(t, spectral_novelty, 'purple')
plt.xlim(30, 40)
plt.xlabel("Time (s)")
plt.title("Spectral-based novelty function")
plt.show()
```
