# Energy-based novelty function
The energy-based novelty function uses root-mean-square energy (RMSE) audio feature. RMSE is a time domain feature that contains information about the overall intensity or strength of an audio signal. This approach is based on the assumption that note onsets correspond with sudden increases in energy.

```python
import numpy as np
from matplotlib import pyplot as plt
import librosa.display
from RhythmRecognition.constants import *
from RhythmRecognition.onset import energy

audiofile = "../audio_files/spark.wav"

energy_novelty = energy.EnergyNovelty(audiofile)
energy_novelty = energy_novelty.get()

frames = np.arange(len(energy_novelty))
t = librosa.frames_to_time(frames, sr=SAMPLING_RATE, hop_length=HOP_LENGTH)

plt.figure(figsize=(12, 3))
plt.plot(t, energy_novelty, 'green')
plt.xlim(30, 40)
plt.xlabel("Time (s)")
plt.title("Energy-based novelty function")
plt.show()
```
