```python
import numpy as np
from matplotlib import pyplot as plt
import librosa.display
import matplotlib as mpl
from RhythmRecognition.constants import *
from RhythmRecognition.onset import rmse

mpl.rcParams.update({'font.size': 16})

audiofile = "../audio_files/vexento.wav"
duration = 60  # how many seconds of the file will be processed

energy_novelty = rmse.EnergyNovelty(audiofile, duration=duration)
energy_novelty = energy_novelty.get()

frames = np.arange(len(energy_novelty))
t = librosa.frames_to_time(frames, sr=SAMPLING_RATE, hop_length=HOP_LENGTH)

plt.figure(figsize=(12, 3))
plt.plot(t, energy_novelty, 'green')
plt.xlim(5, 15)
plt.show()
```
